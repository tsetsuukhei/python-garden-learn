### Арай ахисан түвшин
Давталт ашиглаад бүх файлыг нэгэн зэрэг унших

```python
# бид хичээлээр import pandas гээд pandas.read_csv() гэж ашиглаж байсан бол import pandas as pd
# гэснээр товчлоод ашиглах боломжтой
import pandas as pd
import glob
```

```python
# sales_* гэж бичснээр "sales_" гэж эхэлсэн бүх файлуудыг сонгох болно
files = glob.glob("sales_*.xlsx")

# Бүх файлыг уншаад, нийлүүлэх
# all_data лист дотроо бүх датагаа давталт ашиглаад хялбар байдлаар хадгалах
# давталт нь автоматжуулалтад их хэрэгтэй бөгөөд https://www.youtube.com/watch?v=zmIdC0_0BgY энэ линкээр орж давталт сурч болно.
all_data = []

# Бүх файлыг давтаад листдээ нэмэх
for filename in files:
    df = pd.read_excel(filename)   # эксел файл унших
    all_data.append(df)            # лист дээрээ нэмэх

# Бүх файлыг нийлүүлэх
combined = pd.concat(all_data, ignore_index=True)
```

```python
combined.head()
```

```python
combined.tail()
```

```python
# нийлүүлсэн файлаа хадгалах
combined.to_excel("combined_sales.xlsx", index=False)
```
